# FBS-NK Model One Asset

### Summary of Model

HANK model with Friedman Buffer Stock income process and  wage and price stickiness a la calvo.

The model is heavily drawn from Auclert et al. 2020 ( Adrien Auclert & Matthew Rognlie & Ludwig Straub, 2020. "Micro Jumps, Macro Humps: Monetary Policy and Business Cycles in an Estimated HANK Model," NBER Working Papers 26647, National Bureau of Economic Research, Inc. ) 

Latest version of the paper can be found at https://aauclert.people.stanford.edu/research 



### Model Organization:

Firms
- Final Goods Producer
- Intermediate Goods Producers

Labor Market
- Competitive Labor Packer
- Labor unions

Government

Central Bank


### Indexation notation

Households are indexed by $i \in [0,1]$.

Firms are indexed by $j \in [0,1]$.

Labor types are indexed by $g \in [0,1]$.

nb.: Neither firm nor labor heterogeneity matters for anything in the aggregate.  They are necessary to allow, respectively, sticky prices and sticky wages, but otherwise have little role

## Households

$$
\max_{\{c_{it+s}\}_{s=0}^{\infty}} \mathrm{E_{t}}\left[\sum_{s=0}^{\infty} (\not D \beta_{i})^{t+s} U\left(\mathbf{c_{i t+s}, n_{i t+s}}\right)\right]
$$

Where $U\left(\mathbf{c_{i t}, n_{i t}}\right) = \mathbf{\frac{c_{i t}^{1-\rho}}{1 -\rho} - \varphi \frac{n_{it}^{1+v}}{1+v}}$ 

Households will decide $\mathbf{c_{it}}$ but $\mathbf{n_{it}}$ will be decided by bargaining between a "union" and employers (per ARS)


## Budget Constraint

$$ \pmb{\mathrm{m}}_{i t} = \pmb{\mathrm{z}}_{i t}  + (1+\mathit{r}^{a}_{t})\pmb{\mathrm{a}}_{i t-1}$$

$$ \pmb{\mathrm{c}}_{i t}  + \pmb{\mathrm{a}}_{i t}= \pmb{\mathrm{z}}_{i t}  + (1+\mathit{r}^{a}_{t}) \pmb{\mathrm{a}}_{i t-1}  $$ 

$$\pmb{\mathrm{a}}_{it} \geq 0$$

where 

$\pmb{\mathrm{m}}_{it}$ is cash on hand.

$\pmb{\mathrm{a}}_{it}$ value of mutual fund share holdings.

$\pmb{\mathrm{z}}_{it}$ is labor income.


## Labor Income

Labor income composed of permanent component $\mathit{p}$ and transitory component $\xi$. 


$$\pmb{\mathrm{z}}_{it} = \pmb{\mathrm{p}}_{it} \xi_{it}$$

$$\pmb{\mathrm{p}}_{it}=\pmb{\mathrm{p}}_{i t-1} \psi_{it}$$

$$ (\psi_{it} , \xi_{it}) = \begin{cases}
      (x^{u}_{it}, \mathit{u}) & \ \text{with probability }\mho_{t} \\
      \left(x^{e}_{it}, \frac{(1-\tau_{t})w_{t}N_{t} \theta_{it}}{1-\mho_{t}} \right)  &  \ \text{with probability } 1- \mho_{t} \\
    \end{cases}\\
$$

WLOG assume unemployed individuals lie within $i \in [0, \mho]$

where 

$ log(x^{e}_{it}) \sim N\left( -\frac{\sigma_{\psi}^2}{2} ,  \sigma_{\psi}^2 \right) , \quad \forall i \in [\mho, 1]$
 

$log(x^{u}_{it}) \sim N\left( log(1-k)-\frac{{L\sigma_{\psi}}^2}{2}, L\sigma_{\psi}^2 \right), \quad \forall i \in [0,\mho], \ k \in (0,1),\ L >1 $


$ log(\theta_{it}) \sim N\left(-\frac{\sigma_{\theta}^2}{2}, \sigma_{\theta}^2 \right), \quad \forall i \in [\mho,1]$ 

The idea here is that if/when $(\psi_{it} , \xi_{it})= (x^{u}_{it} , \mathit{u})$ (an individual becomes unemployed), he receives a negative first moment shocks and a positive second moment shock to his permanent income. Markov chain specification for employment is avoided to avoid an additional state variable.



## Law of motion for the Distribution Function

$$ \Phi_{t+1}(\mathbf{a},\mathit{p},\xi) = H\left(\Phi_{t}(\mathbf{a},\mathit{p}, \xi)\right)$$


## Financial Intermediary / Mutual Fund

Beginning of  period outstanding value of assets must equal value of the portfolio of shares and bonds.

$$(1+r^{a}_{t}) A_{t-1} =  B_{t-1} + \int_{0}^{1} (q^{s}_{jt}+ D_{jt})\mathit{v}_{jt-1} \, dj +(1+i_{t-1}) \frac{M_{t-1}}{P_{t}}$$

 Portfolio-Investment Constraint: End of the period t value of newly issued assets equals value of newly purchased shares and bonds. 

$$A_{t} = \frac{M_{t}}{P_{t}} +q^{b}_{t} B_{t} + \int_{0}^{1} q^{s}_{jt}\mathit{v}_{jt}\,dj$$

### Mutual Fund's Problem

$$\max_{\{B_{t}, M_{t} , \mathit{v}_{jt} \}} \mathrm{E}_{t}\left[1+r^{a}_{t+1} \right] = \mathrm{E}\left[ \frac{ B_{t} + \int_{0}^{1} (q^{s}_{jt+1}+ D_{jt+1})\mathit{v}_{jt} \, dj +(1+i_{t}) \frac{M_{t}}{P_{t+1}}}{\frac{M_{t}}{P_{t}} +q^{b}_{t} B_{t} + \int_{0}^{1} q^{s}_{jt}\mathit{v}_{jt}\,dj} \right]$$

 
### FOC: Asset Pricing Equations/ No Arbitrage

$$ \mathrm{E}_{t}\left[1+r^{a}_{t+1}\right]= \frac{1}{q^{b}_{t}}  =\frac{\mathrm{E}_{t}\left[q^{s}_{jt+1} + D_{jt+1} \right]}{q^{s}_{jt}} = (1+i_{t}) \mathrm{E}_{t}\left[\frac{P_{t}}{P_{t+1}}\right] \equiv 1 +r_{t}$$

where 

$A_{t} = \int_{0}^{1} a_{it} \, di$ is the value of assets at the end of period t.

$r_{t}$ is the ex-ante real interest rate.

$i_{t}$ is the nominal interest rate on reserves $M_{t}$.

$B_{t}$ are zero coupon government bonds. Assume $B_{t} =0, \forall t$

$q^{b}_{t}$ is the price of a government bond in period $t$.

$ \mathit{v}_{jt}$ is the portfolio share of firm $j$ stocks where $\int_{0}^{1} \mathit{v}_{jt}\,dj =1$

$ q^{s}_{jt}$ is the stock price of intermediate firm $j$.

$D_{jt} = Y_{jt} -  \frac{W_{t}}{P_{jt}} N_{jt}$ is the dividend of intermediate firm $j$.

$M_{t}$ are nominal reserves at the central bank. Assume $M_{t} = 0 , \forall t$


### Aggregate Stock Market Price

Define $q^{s}_{t} = \int_{0}^{1} q^{s}_{jt}\,dj$ to be the price of the aggregate stock market.

and $D_{t} = \int_{0}^{1} D_{jt} \,dj$ is the aggregate dividend.

Then $$q^{s}_{t} = \frac{\mathrm{E}_{t}\left[q^{s}_{t+1} +D_{t+1}\right]}{1+r_{t}}$$.

## Final Good Producer



Perfectly Competitive Final Good Producer aggregates goods with CES technology

$$ Y_{t} = \left(\int_{0}^{1} Y_{jt}^{\frac{\epsilon_{p}-1}{\epsilon_{p}}}\, dj\right)^{\frac{\epsilon_{p}}{\epsilon_{p}-1}}$$

Final Good Producer Profit maximzation Problem

$$ \max_{Y_{jt}} P_{t} \left(\int_{0}^{1} Y_{jt}^{\frac{\epsilon_{p}-1}{\epsilon_{p}}}\, dj\right)^{\frac{\epsilon_{p}}{\epsilon_{p}-1}} - \int_{0}^{1} P_{jt} Y_{jt} ,\ dj $$


This leads to demand

$$ Y_{jt} = \left(\frac {P_{jt}}{P_{t}}\right)^{- \epsilon_{p}} Y_{t}$$

Plugging this demand into $ P_{t}Y_{t} = \int_{0}^{1} P_{jt} Y_{jt} ,\ dj$ , we obtain

Price index $$P_{t} = \left(\int_{0}^{1} P_{jt}^{1-\epsilon_{p}}\,dj \right )^{\frac{1}{1-\epsilon_{p}}}$$

 ### Firm Maximization Problem
 
 
 Firm $j$ chooses $P_{jt}$ to maximize its dividend $D_{jt}$ and its stock price $q^{s}_{jt} $
 
 $$\max_{\{P_{jt}\}} \overbrace{\frac{(P_{jt} - MC_{t})Y_{jt}}{P_{t}}}^{=D_{jt}} + q^{s}_{jt}\left(P_{jt}\right) $$
 
Given $q^{s}_{jt}\left(P_{jt}\right) = \frac{\mathrm{E}_{t}\left[q^{s}_{jt+1} +D_{jt+1}\left(P_{jt}\right)\right]}{1+r_{t}}$, this is equivalent to: 
 
 $$\max_{\{P_{jt}\}} \mathrm{E}_{t}\left[\sum_{s=0}^{\infty} (\lambda_{P}) ^{s} M_{t,t+s} \left[ \frac{(P_{jt} - MC_{t+s})Y_{jt+s}}{P_{t+s}}\right]\right]$$
 
subject to $$Y_{jt} = \left(\frac {P_{jt}}{P_{t}}\right)^{- \epsilon_{p}} Y_{t}$$
 
where $ \lambda_{P}$ is the  probability a firm cannot change its price,  $M_{t, t+s} = \prod_{k=t}^{t+s-1} \frac{1}{1+r_{k}}$ is the stochastic discount factor and $MC_{t} = \frac{W_{t}}{A_{t}}$ is the marginal cost

### The optimal price :
For firm $j$

$$ P_{jt}^{*} = \frac{\epsilon_{p}}{\epsilon_{p}-1} \frac{\sum_{s=0}^{\infty}\lambda_{p}^{s} \mathrm{E}_{t}\left[M_{t,t+s} \frac{Y_{t+s}}{P_{t+s}^{1-\epsilon_{p}}} MC_{t+s}\right]}{\sum_{s=0}^{\infty}\lambda_{p}^{s} \mathrm{E}_{t}\left[M_{t,t+s} \frac{Y_{t+s}}{P_{t+s}^{1-\epsilon_{p}}}\right]}$$


Taking logs , optimal price can be expressed as 

$$ p_{t}^{*} = ln\left(\frac{\epsilon_{p}}{\epsilon_{p}-1}\right) + \left(1 - \frac{\lambda_{p}}{1+r^{*}}\right) \sum_{s=0}^{\infty} \lambda_{p}^{s} \mathrm{E}_{t} \left [ M_{t,t+s} [w_{t+s} - a_{t+s}]\right]$$

where the variables above are logged and $r^{*}$ is the steady state interest rate. 


 ## Phillips Curve:

$$ \pi_{t} = \frac{\mathrm{E}_{t}[\pi_{t+1}]}{1+r^{*}} + \lambda (\mu_{t}-\mu)$$

where $r^{*}$ is the natural rate of interest in the steady state, $\lambda = \frac{(1-\lambda_{p})(1-\frac{\lambda_{p}}{1+r^{*}})}{\lambda_{p}}$,  $ \mu_{t} = log(P_{t}) - log(W_{t}) + log(A_{t})$ and $\mu = \frac{\epsilon_{p}}{1-\epsilon_{p}}$

## Labor Market

Every worker $i$ provides $n_{igt}$ hours of work to labor union $g \in [0,1]$ and assume $n_{igt} = \mathit{n}_{gt}$. This assumption will imply labor income heterogeneity is only due to transitory income shocks.

Therefore, 

$$n_{it} = \int_{0}^{1} n_{igt}\,dg$$ Should the bounds of the integral be from $ \mho$ to $1$

and 

$$N_{gt} = \int_{0}^{1} n_{igt}\,di$$ 

OR
$$N_{gt} = \int_{\mho}^{1} n_{igt}\,di = (1-\mho_{t}) \mathit{n}_{gt}$$ 

Can think of LHS as labor demand for labor type $g$ o, and RHS is effective labor supply of labor type $g$ from households. There is an underlying assumption that unemployed households supply  "useless" labor. That they're labor supply is effectively useless. Or assumption can be thought as labor union only asks for labor from unemployed households. So if labor unions need to supply $N_{gt}$ to the labor packer, then the labor union will demand $ n_{gt} = \frac{N_{gt}}{1-\mho}$ from each household that is working.


## Competitve Labor  Packer

Perfectly Competitive Labor Packer purchases labor from Labor Unions and  aggregates Labor using CES technology and sells $N_{t}$ to firms at price $W_{t}$


$$ N_{t} = \left(\int_{0}^{1} N_{gt}^{\frac{\epsilon_{w}-1}{\epsilon_{w}}}\,dg\right)^{\frac{\epsilon_{w}}{\epsilon_{w}-1}}$$

The Competitve Labor Packer's profit maximizing Problem 

$$ \max_{n_{jgt}} W_{t} \left(\int_{0}^{1} N_{jgt}^{\frac{\epsilon_{w}-1}{\epsilon_{w}}} \, dg \right)^ {\frac{\epsilon_{w}}{\epsilon_{w}-1}} - \int_{0}^{1} W_{gt}N_{jgt}\, dj $$


### Competitive Labor Packer Demand for labor types

$$ N_{gt} = \left(\frac{W_{gt}}{W_{t}}\right)^{-\epsilon_{w}} N_{t} $$

Wage index follows
$$ W_{t} = \left(\int_{0}^{1} W_{gt}^{1-\epsilon_{w}}\,dg\right)^{\frac{1}{1-\epsilon_{w}}}$$


## Labor Union Maximization Problem

Labor Union $g$ will set a wage to maximize expected lifetime utility. It may only adjust its price given it is chosen by the calvo fairy. 

$$ \max_{\{W_{gt}\}} \mathrm{E_{t}}\left[\sum_{s=0}^{\infty} (\not D \lambda_{w})^{s} \int_{0}^{1} \beta_{i}^{t+s} U\left (c(W_{t+s}), n_{i t+s}) \, d \Phi_{it+s} \right)\right] $$

subject to the following three constraints $$ N_{gt} = \left(\frac{W_{gt}}{W_{t}}\right)^{-\epsilon_{w}} N_{t} $$

$$ W_{t} = \left(\int_{0}^{1} W_{gt}^{1-\epsilon_{w}}\,dg\right)^{\frac{1}{1-\epsilon_{w}}}$$

where $\lambda_{w}$ probability labor union cannot adjust its wage and $\Phi_{it+s}$ is the distribution of  liquid assets, transitory and permanent shocks over households at period $t+s$.

## Optimal Wage

$$ W_{t}^{*} = \frac{\epsilon_{w}}{\epsilon_{w}-1} \frac{\mathrm{E}_{t} \left[\sum_{s=0}^{\infty} (\not D \lambda_{w})^{s} N_{t+s} MV_{t+s} W_{t+s}^{\epsilon_{w}} \right]} {\mathrm{E}_{t} \left[ \sum_{s=0}^{\infty} (\not D \lambda_{w})^{s} \frac{(1-\tau_{t+s})}{P_{t+s}} N_{t+s}  MU_{t+s} W_{t+s}^{\epsilon_{w}}\right]}$$

where $MU_{t+s} = \int_{0}^{1}  \beta_{i}^{t+s} \theta_{it} U_{c} \left(c_{i t+s}, n_{i t+s} \right) \  di $

$ \quad \quad MV_{t+s} = - \int_{0}^{1}  \beta_{i}^{t+s} U_{n} \left(c_{i t+s}, n_{i t+s} \right) \ di $

Taking the log of both sides and isolating for $w$

$$ w_{t}^{*} = w - \tau - p - mrs + (1 - \not D \lambda_{w}) \mathrm{E}_{t}\left[\sum_{s=0}^{\infty} ( \not D \lambda_{w})^{s} ( mrs_{t+s} + p_{t+s} + \tau_{t+s})\right]$$

$$ w_{t}^{*} = ln\left(\frac{\epsilon_{w}}{\epsilon_{w}-1}\right) + (1 - \not D \lambda_{w}) \mathrm{E}_{t}\left[\sum_{s=0}^{\infty} ( \not D \lambda_{w})^{s} ( mrs_{t+s} + p_{t+s} + \tau_{t+s})\right]$$

where $w , \tau, p$ are logged steady state values, $mrs_{t+s} = log \left(- \frac{\int_{0}^{1}  \beta_{i}^{t+s} U_{n} \left(c_{i t+s}, n_{i t+s} \right) \ d \Phi_{it+s} ) }{\int_{0}^{1}  \beta_{i}^{t+s} \theta_{it} U_{c} \left(c_{i t+s}, n_{i t+s} \right) \  d \Phi_{it+s} )} \right)$

Note $w- \tau - p - mrs$ is the markup of the wage over the flexible equilibrium level.

$$ \pi_{t}^{w} =  \not D  \mathrm{E}_{t} \left[ \pi_{t+1}^{w}\right] + \frac{(1-\lambda_{w})}{\lambda_{w}} (1-\not D \lambda_{w}) (\mu^{w} - \mu_{t}^{w})$$

# Boppart, Krusell, and Mitman Algorithm

1. Obtain steady state.

2. Set a path for the exogenous shock $(Z_{t} , v_{t})_{t=0}^{t=T}$, and guess the path of $(Y_{t})_{t=0}^{t=T}$. This guess gives the real wage, labor supply. With an Terminal inflation $\pi_{T}$ we can obtain $\pi_{T-1}$ which in turn gives us $\pi_{T-1}^{w}$.  With inflation and output we can now determine the nominal interest rate which in turn gives us the real interest rate.  

How do we determine aggregate assets? We have $D_{T}$ and $q_{T}$. Since we have path of interest rates we can then figure out the rest of the stock prices $(q_{t})_{t=0}^{t=T}$. 

3. With real wage,  real interest rate, and the labor supply,  we can compute the optimal policies using EGM. 

4. Starting at the steady state in period $t=0$, Simulate the model with the policies solved previously with only idiosyncratic shocks (with no aggregate shocks as they are already imputed within the paths of the interest rate and the wage).

5. Determine Whether goods market clears, and whether wage inflation equation clears as the RHS of the wage inflation equation takes in each agent's consumption.

6. If the distance between the guess of the path of output and aggregate consumption obtained from the simulation is too large, then we will need ot rerun the algorithm.

7. Update guesses by taking complex combination between Consumption path generated from simulation , and initial output guess.

## Krusell Smith Algorithm

Three Forecasting rules:

A rule for inflation

A rule for expected inflation

A rule for aggregate assets.


A rule for inflation and expected inflation pins down the markup, with $Z_{t}$ as given it will then pin down the real wage.

With the real wage, we can then pin down the labor supply, which in turn pins down output.

With output, we can then pin down the nominal interest rate.

With the nominal interest rate and expected inflation we can then pin down the real interest rate. 

Thus now we have the real wage and the interest rate tomorrow to determine the consumption decision.

In addition, with the rule for aggregate assets tomorrow, which has to equal the stock price tomorrow,  with the dividend tomorrow pinned down from real wage, labor supply and output, and finally with the interest rate tomorrow we can pin down the stock price today.

## What are the main endogenous variables that determine all other endogenous variables?

Exogenous Variables: $Z_{t} , v_{t}$

Main Endogenous Variables: $w_{t}, \pi_{t}$

If I have the real wage, I can determine Labor  from the budget constraint, which then allows me to determine Output.

If I have inflation I can determine the nominal rate from the taylor rule, expected inflation form the phillips curve and the real rate. 

If I have inflation and the real wage I can determine wage inflation, then I can determine expected wage inflation.

With Labor , I can determine output, given that output must equal consumption, and given aggregate resources today (initial level of aggregate resources) I can determine assets which in turn tells me the stock price today. 

The real wage, consumption, and output should give me wage inflation.




### How do we determine aggregate variables when simulating productivity shocks?

Suppose there is a aggregate productivity shock and suppose households follow rules to determine their saving policies so we have the path of aggregate asset demand. Aggregate assets with the current state of aggregate resources gives us aggregate consumption. With aggregate consumption, we haev aggregate output and thus aggregate labor. WIth the Labor we have the real wage.

With the real wage we can obtain, the price and wage index, the initial price and wage, and the the fact that optimal $\frac{p^{*}}{w^{*}} = \mu_{p}$ ,we can figure out the optimal price and optimal wage individuals (this may require root finding). With this, we can now figure out inflation and wage inflation which will give us the nominal rate and expected inflation which tells us the real rate. 

### How do we determine aggregate variables when simulating monetary policy shock?

Again we assume we have aggregate assets today, and thus aggregate consumption, once again this tells us output, labor and wage , which tells us inflation, but now the $v$ term in the taylor rule has changed, so the real interest rate will be different.
